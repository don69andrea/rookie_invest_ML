## Executive Summary

Ziel dieses Projekts ist es, die Wahrscheinlichkeit eines späteren
Formel-1-Einstiegs auf Basis der Performance von Fahrern in F2 und F3 vorherzusagen.

Hierzu wurde ein erklärbares Machine-Learning-Modell entwickelt,
das Fahrer nach ihrem relativen F1-Potenzial priorisiert.
Die Modellierung berücksichtigt sportliche Leistung,
Konstanz, Rennerfahrung und Pace-Indikatoren.

Die Ergebnisse zeigen, dass ein signifikanter Anteil der späteren
F1-Einsteiger bereits in den Top 10–15 % der vom Modell priorisierten Fahrer enthalten ist.
Dies macht den Ansatz besonders geeignet für Scouting- oder Investment-Entscheidungen.

Ein Vergleich mit einem Random-Forest-Modell bestätigt die Wettbewerbsfähigkeit
der logistischen Regression, welche bewusst als interpretierbares Baseline-Modell gewählt wurde.

Das Modell dient nicht als deterministische Karriereprognose,
sondern als datenbasierte Entscheidungsunterstützung.


## Limitationen und mögliche Verzerrungen

Trotz sorgfältiger Modellierung weist die Analyse mehrere Limitationen auf:

- **Zeitverzögerte Zielvariable**  
  Ein F1-Einstieg erfolgt oft mehrere Jahre nach den beobachteten F2/F3-Saisons.
  Für jüngere Jahre sind daher noch nicht alle Zielwerte beobachtbar.

- **Historische Verzerrung**  
  Die Struktur von Nachwuchsserien und der Weg in die Formel 1
  haben sich über die Jahre verändert, was die Vergleichbarkeit einschränkt.

- **Selektionsbias**  
  Die Analyse berücksichtigt nur Fahrer mit dokumentierten Karrieredaten,
  wodurch bestimmte Gruppen unterrepräsentiert sein können.

- **Korrelation zwischen Features**  
  Viele Performance-Merkmale sind stark miteinander korreliert,
  was die Interpretation einzelner Koeffizienten erschwert.

Diese Limitationen werden bewusst in Kauf genommen,
da das Ziel des Projekts eine realistische Priorisierung
und keine exakte Karriereprognose ist.
